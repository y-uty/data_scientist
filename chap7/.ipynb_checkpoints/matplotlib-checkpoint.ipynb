```python
#データの可視化　⇒　データ分析結果を説明する上で重要
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline
%precision 3
```

```python
#棒グラフ
#表示するデータ
x = [1,2,3] #棒グラフは1次元なので、これはただのデータの通し番号
y = [10,1,4]
#グラフの大きさ指定
plt.figure(figsize = (10,6))
#棒グラフ
plt.bar(x, y, align='center', width=0.5)
#各データのラベル
plt.xticks(x,['A Class', 'B Class', 'C Class'])
#x,yのラベルを設定
plt.xlabel('Class')
plt.ylabel('Score')
#グリッドを表示
plt.grid(True)
```

```python
#横の棒グラフ
x = [1,2,3] #棒グラフは1次元なので、これはただのデータの通し番号
y = [10,1,4]
plt.figure(figsize = (10,6))
plt.barh(x, y, align='center')
plt.yticks(x,['A Class', 'B Class', 'C Class'])
plt.ylabel('Class')
plt.xlabel('Score')
plt.grid(True)
```

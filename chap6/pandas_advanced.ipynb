```python
#データの結合
#データ1の準備
data1 = {
    'id':['100','101','102','103','104','106','108','110','111','113'],
    'city':['Tokyo','Osaka','Kyoto','Hokkaido','Tokyo','Tokyo','Osaka','Kyoto','Hokkaido','Tokyo'],
    'birth_year':[1990,1989,1992,1997,1982,1991,1988,1990,1995,1981],
    'name':['Hiroshi','Akiko','Yuki','Satoru','Steeve','Mituru','Aoi','Tarou','Suguru','Mitsuo']
}
df1 = DataFrame(data1)
df1
```

```python
#データ2の準備
data2 = {
    'id':['100','101','102','105','107'],
    'math':[50,43,33,76,98],
    'english':[90,30,20,50,30],
    'sex':['M','F','F','M','M'],
    'index_num':[0,1,2,3,4]
}
df2 = DataFrame(data2)
df2
```

```python
#mergeメソッド
#内部結合(デフォルト)
print('★内部結合')
pd.merge(df1, df2, on = 'id') #onパラメータで結合キーを指定する
```

```python
#全結合
print('★全結合')
pd.merge(df1, df2, on = 'id', how = 'outer') #howパラメータで結合方法を指定
#値が無い箇所は NaN になる( Null のようなもの)
```

```python
#左外部結合
print('★左外部結合')
pd.merge(df1, df2, on = 'id', how = 'left') # 1つめのテーブルが主体
```

```python
#右外部結合
print('★右外部結合')
pd.merge(df1, df2, on = 'id', how = 'right') # 2つめのテーブルが主体
```

```python
#マッピング処理 ⇒ Excelのvlookupに近い処理
#参照データ
city_map = {
    'Tokyo':'Kanto',
    'Hokkaido':'Hokkaido',
    'Osaka':'Kansai',
    'Kyoto':'Kansai'
}
city_map
```

```python
#元となるデータから、参照データの対応するデータを持ってきて、新しくカラムとして表示する
df1
```

```python
df1['region'] = df1['city'].map(city_map)
df1
```

```python
df1['city_str'] = df1['city'].map(lambda x: str(x)[0:3])
df1
```

```python
#DataFrameから列を削除
# df1 = df1.drop(['city_str','region_str'], axis=1)
```

```python
#ビン分割　⇒　離散的な範囲でデータを分割・集計する場合に使う
#cut関数　分割の粒度を定義しておいて、DataFrameのカラムと共に引数に指定する
year_bins = [1980,1985,1990,1995,2000]
birth_year_cut_data = pd.cut(df1.birth_year, year_bins)
birth_year_cut_data
```

```python
#value_counts関数　cut関数の結果でそれぞれの数を集計する
pd.value_counts(birth_year_cut_data)
```

```python
#cut関数のlabelsパラメータでビンに名前を付けることができる
grp_nm = ['early1980s','late1980s','early1990s','late1990s']
birth_year_cut_data = pd.cut(df1.birth_year, year_bins, labels=grp_nm)
pd.value_counts(birth_year_cut_data)
```

```python
#cut関数　ビンのリストを用意するのではなく、予め分割数を決めておくことも可能
#ただし、データによっては割り切れず小数点以下が出るので注意
birth_year_cut_data_2 = pd.cut(df1.birth_year, 5)
pd.value_counts(birth_year_cut_data_2)
```

```python
#qcut関数　x分位点での分割を、xのパラメータ指定で実現する (qはquantilesのこと)
#四分位点分割
birth_year_cut_data_3 = pd.qcut(df1.birth_year, 4)
pd.value_counts(birth_year_cut_data_3)
```

```python
#サイズ情報　sqlでいうcount関数
df1.groupby('city').size()
```

```python
#平均
df1.groupby('city')['birth_year'].mean()
```

```python
#groupbyメソッドのas_indexパラメータをFalse指定すると、インデックスがなくなりそのままテーブルとして使える
df1.groupby(['region','city'], as_index=False)['birth_year'].mean()
```

```python
#イテレータ　groupbyメソッドで、反復的に値を取り出す機能　forループなどで処理できる
#ex.)group 集計軸の名前取り出し　　subdf 集計軸の各値に合致する行のみをすべて抽出
for group, subdf in df1.groupby('region'):
    print('===========================================')
    print('リージョンネーム：{0}'.format(group))
    print(subdf)
```

```python
#複数の計算をまとめて行いたい場合は、aggメソッドの引数に実行したい関数名のリストを渡して使うと便利
mat2 = pd.read_csv('student-mat.csv', sep=';')
func_list = ['count','mean','max','min']
mat_grouped = mat2.groupby(['sex','address'])
mat_grouped['age','G1','G2','absences'].agg(func_list)
```

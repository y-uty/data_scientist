```python
#必要なライブラリのインポート
import numpy as np
import numpy.random as random
from numpy import nan as NA
import scipy as sp
import pandas as pd
from pandas import Series #1次元配列
from pandas import DataFrame #2次元配列
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%precision 3
random.seed(0)
```

```python
#欠損データの生成
df = pd.DataFrame(np.random.rand(10,4))
df.iloc[1,0] = NA
df.iloc[2:3,2] = NA
df.iloc[5:,3] = NA
df
```

```python
#リストワイズ削除 dropnaメソッドを適用　NaNがある行をすべて取り除くが、データが極端に少なくなる可能性
df.dropna()
```

```python
#ペアワイズ削除 使いたい列を取り出してから、dropnaメソッドを適用
df[[0,1]].dropna()
```

```python
#fillnaメソッドで埋める　引数に代替する値を渡す　ex.)df.fillna(0)
df.fillna(0)
```

```python
#前の値で埋める　時系列データの処理で便利
df.fillna(method = 'ffill')
```

```python
#各カラムの平均値で埋める(平均値代入法)　時系列を扱うときは過去の欠損値を未来データを使って埋めることになるので注意
df.mean()
```

```python
df.fillna(df.mean())
```

```python
#pandas-datareaderをインストールしてから、ライブラリをインポート
import pandas_datareader.data as pdr
```

```python
#2001年から2016年までのドル円為替レートデータ(DEXJPUS)を使う
stt_dt = '2001/01/01'
end_dt = '2016/12/31'
fx_jpusdata = pdr.DataReader('DEXJPUS', 'fred', stt_dt, end_dt)
#土日はレコードなし、祝日はレコードはあるが値がNaN
```

```python
fx_jpusdata.head()
```

```python
#特定の年月のデータを参照する
fx_jpusdata['2016-04']
```

```python
fx_jpusdata.resample('M').last()
```

```python
fx_jpusdata.resample('Y').last()
```

```python
#欠損している休日分のデータを埋める
fx_jpusdata.resample('D').last().head(10)
```

```python
fx_jpusdata.resample('D').ffill().head(10)
```

```python
#比率の計算　⇒　データをずらして、前日比を簡単に計算する
fx_jpusdata.shift(1).head() #DATEのインデックスはそのままで、データが1日ずつ後ろにずれる(前日分のデータが入る)
```

```python
#1日ずれのデータを使って前日比を計算する
fx_jpusdata_fill = fx_jpusdata.resample('D').last().ffill()
fx_jpusdata_ratio = fx_jpusdata_fill / fx_jpusdata_fill.shift(1)
fx_jpusdata_ratio.head(15)
```

```python
fx_jpusdata_fill.head(15)
```

```python
#移動平均　rollingメソッド
fx_jpusdata.rolling(3).mean().head(30) #3日間の移動平均
```

```python
fx_jpusdata.rolling(3).std().head(30) #3日間の標準偏差
```
